```python
train["gender"] = train["gender"].astype('category')
train["topic"] = train["topic"].astype('category')
train["sign"] = train["sign"].astype('category')
train["user.id"] = train["user.id"].astype('category')
train.dtypes
```

```python
# remove duplicate spaces
train['text'] = train['text'].apply(lambda x: re.sub(' +', ' ',str(x)))
# strip spaces around the sentence string
train['text'] = train['text'].apply(lambda x: str(x).strip())
# get the word count for each blog
train['word_count'] = train['text'].apply(lambda x: len(str(x).split(" ")))
```

```python
train.head()
```

```python
# Average word length
def avg_word(sentence):
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))

train['avg_word'] = train['text'].apply(lambda x: avg_word(x))
train[['text','avg_word']].head()
```

```python
# Number of numerics appeared in each blog
train['numerics'] = train['text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
train[['text','numerics','age']].head(20)
```

```python
# Number of hashtags
train['hastags'] = train['text'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
train[['text','hastags','age']].head(50)
```

```python
# Change upper case characters into lower cases.
train['text'] = train['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
train['text'].head()
```

```python
# Removal of stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')
train['text'] = train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
train['text'].head()
```

```python
# Find the 50 most common words
freq = pd.Series(' '.join(train['text']).split()).value_counts()[61:100]
freq
```
